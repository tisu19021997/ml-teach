# while

* Every loop MUST have a stop condition. 
* Else, it runs forever.

```python
i = 3
while i != 0:
    i = i - 1
    print("meow")
    
# i = 3
# print => loop1 i = 2
# print => loop2 i = 1
# print => loop3 i = 0


# i = 3
# i = 2 => print 
# i = 1 => print 
# i = 0 => print 
```

```python
i = 3
while True:
    i = i - 1
    
    if i == 0:
        break
    else:
        print('meow')
        continue
        print('meow meow')
```

```python
data = [1, 2, 3, 4, 5, 6]
i = 0
while True:
    if i == len(data) - 1:
        break
    if i % 2 != 0:
        i += 1
        continue

    print(data[i], data[i]**2)
    i += 1
```

## list comprehension

* Another way to quickly loop through list is to use list comprehension

```python
list1 = [0, 1, 2, 3, 4]
list2 = [value**2 for value in list1]

print(list2)
```

* List comprehension with index

```python
list1 = [0, 1, 2, 3, 4]
list2 = [(idx, x**2) for idx, x in enumerate(list1)]

print(list2)
```

* Add a layer of conditions

```python
list1 = [0, 1, 2, 3, 4]
list2 = [
    x**2 if i % 2 == 0 else x
    for i, x in enumerate(list1)]

print(list2)

# if i % 2 == 0: return x**2
# else: return x
```

* Write a function to print a square using # of size n * m

```python
def print_square(n, m):
    pass
```

*click to show solution*

```python
def print_square(n, m):
    for row in range(n):
        char = '#'
        for col in range(m):
            if row == 0 or row == n - 1 or col == m - 1:
                char += '#'
            else:
                char += ' '
        print(char)

print_square(12, 12)
```

* Write a function to take a word and a sentence contain that word multiple times as parameters and make that word uppercase.
    * `your_function("dang", "minh dang la con cho, minh dang co dang nguoi mau")` => "minh Dang la con cho, minh Dang co Dang nguoi mau"

```python
def your_function(word, sentence):
    pass
```

*click to show solution*

```python
def your_function(word, sentence):
    new_sentence = []
    for w in sentence.split():
        if w == word:
            w = w.title()
        new_sentence.append(w)
    return ' '.join(new_sentence)

your_function("dang", "minh dang la con cho, minh dang co dang nguoi mau")
```
