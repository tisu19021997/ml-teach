* Write a program to return the BMI status given the user's weight and height

*click to show answer*

```python
def bmi_status(weight, height):
    bmi = weight / (height * height)
    if bmi >= 40.0:
        return 'Obese'
    elif bmi >= 25:
        return 'Overweight'
    elif bmi >= 18.5:
        return 'Normal'
    else:
        return 'Underweight'
    
assert bmi_status(70, 1.7) == 'Normal'
assert bmi_status(70, 1.5) == 'Overweight'
assert bmi_status(70, 1.3) == 'Obese'
assert bmi_status(40, 1.7) == 'Underweight'
```
