## Element navigation

```python
from PIL import Image
import requests
import IPython
```

**Select using `driver.find_element(By.CSS_SELECTOR, element_class)`**

```python
img_class = 'picture'
logo = driver.find_element(By.CSS_SELECTOR, img_class)

# get_attribute
logo_html = logo.get_attribute('innerHTML').strip()
print(logo_html)
```

**Try to get the image src (link) using regex**

```python
# Solution
img_src_regex = r'<img .* src="(.+[.png|.jpg|.webm])"\s'
logo_src = re.findall(img_src_regex, logo_html)
print(logo_src)
```

* Kết quả trả lẫn quá nhiều html tag, nếu viết regex thì quá khó.
* Sử dụng thư viện beautifulsoup: `pip install beautifulsoup4` 

```python
# Clear html tags using BeautifulSoup.
content_class = '.mw-parser-output > p'
contents = driver.find_elements(By.CSS_SELECTOR, content_class)

for content in contents:
    content_text = bs.BeautifulSoup(
        content.get_attribute('innerHTML')).get_text()
    print(content_text)
```
